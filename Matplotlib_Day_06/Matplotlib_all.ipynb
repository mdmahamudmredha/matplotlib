## matplotlib Day 05

```python
x = [1, 2, 3, 4]
y1 = [10, 20, 15, 25]
y2 = [12, 18, 22, 28]

plt.plot(x,y1,color=(1,0,0),ls='-.', lw='5',marker='d',mfc='g', ms=15, mec='b', alpha=1, zorder=2 )
plt.plot(x,y2, color=(0,1,0),ls='-.', lw='5',marker='d',mfc='g', ms=15, mec='b', alpha=1, zorder=3 )
plt.title("Yearly Sales Comp:")
plt.xlabel('Quarter')
plt.ylabel("Sales")
plt.show()
```

```python
x = [1, 2, 3, 4]
y1 = [10, 20, 15, 25]
y2 = [12, 18, 22, 28]

plt.plot(x,y1,'r--d')
plt.plot(x,y2, color=(0,1,0),ls='-.', lw='5',marker='d',mfc='g', ms=15, mec='b', alpha=1, zorder=3 )
plt.title("Yearly Sales Comp:")
plt.xlabel('Quarter')
plt.ylabel("Sales")
plt.show()
```

## matplotlib Day 06

```python
#import matplotlib.pyplot as plt

study_hours = [1, 2, 3, 4, 5, 6, 7] 
exam_score = [40, 45, 50, 55, 60, 66, 70]
plt.xlabel("study_hours")
plt.ylabel("exam_score")
plt.grid()
plt.scatter(study_hours,exam_score);
```

```python

study_hours = [1, 2, 3, 4, 5, 6, 7]

score_A = [40, 45, 50, 55, 60, 66, 70]
score_B = [38, 42, 48, 52, 56, 61, 67]

plt.scatter(study_hours, score_A, color='blue', marker='o', label='Student A')
plt.scatter(study_hours, score_B, color='red', marker='x', label='Student B')

plt.title("Study Hours vs Exam Score (Two Students)")
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.show()
```

```python
price = [
    130, 400, 320, 110, 360, 500, 420, 1200, 520, 125,
    1300, 145, 520, 165, 1400, 140, 480, 85, 115, 150,
    100, 1000, 105, 95, 175, 300, 135, 80, 120, 90,
    155, 140, 170, 440, 340, 1100, 150, 460, 160, 470
]

sales = [
    485, 250, 290, 505, 270, 200, 240, 60, 190, 490,
    50, 470, 190, 450, 30, 475, 210, 530, 500, 465,
    515, 100, 510, 520, 440, 300, 480, 540, 495, 525,
    460, 475, 445, 230, 280, 80, 465, 220, 455, 470
]

plt.figure(figsize=(10, 6))
plt.scatter(price, sales, color='teal', alpha=0.8, s=100, edgecolors='black')
plt.title("Product Price vs Number of Sales (E-commerce Clustering Example)")
plt.xlabel("Product Price (in Taka)")
plt.ylabel("Number of Sales")
plt.grid(True)
plt.show()
```
